langchain支持大部分常见大语言模型LLM，并提供了对应的依赖库

gpt:langchain-openai

claude:langchain-anthropic

deepseek:langchain-deepseek

gemini:langchain-google-genai

ollama:langchain-ollama



## 1.开发步骤
1.引入依赖

- uv add langchain

- uv add langchian-deepseek

2.配置环境:配置api_key

3.初始化模型（两种方法）

- 使用init_chat_model函数，由langchain自动创建模型对象
- 使用不同模型对应的Model类，手动创建模型对象


## 2.初始化模型

### 2.1.方式一使用init_chat_model
- 最简单的方式

1.引入init_chat_model函数   
2.读取环境变量  
3.直接用模型名字调用即可



```python
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv

load_dotenv()

model=init_chat_model("deepseek-chat")

print(type(model))
```

### 2.2.方式二使用init_chat_model+自定义模型
1.配置环境变量
- api_key
- base_url  
2.读取环境变量  
3.创建模型
- 名字
- 提供商
- api_key
- base_url

```python
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv
import os    # 加载环境变量

load_dotenv()

api_key=os.getenv("DEEPSEEK_API_KEY")
base_url=os.getenv("BASE_URL")

model=init_chat_model(
    model="qwen-max",
    model_provider="openai",
    api_key=api_key,
    base_url=base_url
)

print(type(model))
```

### 2.3.方式三使用社区版本Model类来创建
以社区版本Model类来访问阿里云百炼的模型。   
1.首先安装langchian的社区依赖   
uv add langchain-community    
2.安装阿里云百炼依赖    
uv add dashscope  
3.然后就可以使用Model类来创建模型了

```python
from langchain_community.chat_models import ChatTongyi
from dotenv import load_dotenv

load_dotenv()

model=ChatTongyi(
    model="qwen-max",
)

print(type(model))
```
