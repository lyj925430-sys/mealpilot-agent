## 1.创建Agent
langchain提供了create_agent方法来创建Agent。    
当我们创建Agent时，可以使用我们事先创建好的模型，也可以通过模型名称直接创建（这种只针对于可以使用init_chat_model创建的模型有效）

### 1.1.方式一：使用模型名称直接创建Agent

```python
from langchain.agents import create_agent

agent = create_agent("deepseek-chat")
```

### 1.2.方式二：使用事先创建好的模型

```python
#创建模型
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv
import os 

load_dotenv()

api_key=os.getenv("DASHSCOPE_API_KEY")
base_url=os.getenv("BASE_URL")

model=init_chat_model(
    model="qwen-max",
    model_provider="openai",
    api_key=api_key,
    base_url=base_url
)
```

```python
#创建Agent
from langchain.agents import create_agent

agent=create_agent(
    model=model
)
```

## 2.调用Agent
智能体也分为阻塞式调用和流式调用

### 2.1.方式一invoke:阻塞式调用
传的是一个对象  
{
    "messages":[
        {"role":"user","content":"你是谁？"}
    ]
}

因为之后还会有AI的信息、工具的信息等，所以用一个对象封装，方便迭代。

```python
response = agent.invoke({
    "messages":[
        {"role":"user","content":"你是谁？"}
    ]
})
print(response)
```

### 2.2.方式二stream：流式调用

```python
response=agent.stream({
    "messages":[
        {"role":"user","content":"你是谁？?"}
    ]
},
stream_mode="messages"  #流式的模式：信息片段
)

for token,metadata in response:
    if token.content:
        print(token.content,end="",flush=True)
```
