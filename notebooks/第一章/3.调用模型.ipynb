先初始化模型，我这里使用init_chat_model()方法+自定义模型

```python
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv
import os

load_dotenv()

api_key=os.getenv("DASHSCOPE_API_KEY")
base_url=os.getenv("BASE_URL")

model=init_chat_model(
    model="qwen-max",
    model_provider="openai",
    api_key=api_key,
    base_url=base_url,
)
print(type(model))
```

## 1.方式一invoke:阻塞式调用
需要等待模型生成全部结果才会返回，等待时间长

```python
response=model.invoke("你是谁？")
print(response)
```

## 2.方式二stream：流式调用
实时看到AI返回的一个个词  
model.stream() 只负责 “建立连接”，真正发送请求、获取数据、调用模型，是在 for 循环开始迭代的时候才发生的。

```python
response=model.stream("你是谁？")
for chunk in response:
    print(chunk)
```
