## 1.LangSmith
- Langchain是基于LangGraph的，而LangGraph提供了完整的后端部署功能、API接口
- LangGraph还提供了基于LangSmith的GUI控制台实现Agent的调试、部署、监控

### 1.1.配置LangSmith
- LangSmith提供了对Agent的GUI管理界面   
- 首先，我们要注册LangSmith，开通服务，生成API_KEY，配置到.env里  

### 1.2.开发Agent
- 首先，我们把刚才的代码集中到一个py文件中
- LangGraph会自动托管Agent的记忆，因此代码中不用自己添加checkpointer！
- LangGraph自带Restful的API接口，我们只要定义好Agent就可以，其它不用管

### 1.3.本地部署
- 使用LangGraph命令行在本地部署，所以要先安装LangGraph的依赖：uv add langgraph-cli[inmem]
- 然后，在项目根目录添加一个langgraph配置文件
- 注意：其中的agent配置格式为：[Agent文件路径]:[Agent变量名]
- 使用LangGraph命令本地部署Agent：uv run langgraph dev
