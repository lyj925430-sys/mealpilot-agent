## 1.uv是什么
uv = 极速 + 全能的 Python 包管理 + 虚拟环境 + 项目管理工具，一个顶一堆传统工具


使用uv创建项目的同时，它会给我们创建一个独立的环境，避免和其他项目环境冲突

## 2.安装uv
1.安装：按下 Win + R 键，输入 powershell 并回车，打开 PowerShell 窗口

  ①运行（清华源）：pip install uv -i https://pypi.tuna.tsinghua.edu.cn/simple

  ②检查是否安装成功：uv --version。看到类似 uv 0.x.x 的字样，说明 uv 已经就绪。

2.添加镜像源：为了以后用 uv 安装任何包都极速，我们需要把清华源设为默认.在 PowerShell 中直接运行以下命令（它会帮你自动创建一个配置文件）：

$configDir = "$env:APPDATA\uv"; if (!(Test-Path $configDir)) { New-Item -ItemType Directory -Path $configDir }; "[[index]]`nurl = `"https://pypi.tuna.tsinghua.edu.cn/simple`"`ndefault = true" | Out-File -FilePath "$configDir\uv.toml" -Encoding utf8

## 3.创建项目（vscode）
1.准备项目文件夹：打开 VS Code，点击 “打开文件夹”，新建一个文件夹并进入。

2.初始化项目：打开 VS Code 内置的终端（快捷键 Ctrl + ~），输入以下命令：uv init

3.创建虚拟环境：在终端输入：uv venv --python 3.13.3。python版本可以通过python --version查询

4.关联 VS Code

    1.按快捷键 Ctrl + Shift + P
    2.输入 Python: Select Interpreter 并回车
    3.选择标有 Python 3.1x.x ('.venv': venv) 的那一项

## 4.可以选择使用jupyter
需要在项目中引入notebook依赖：uv add notebook
